# pyplot 

Each pyplot function make some changes to a figure e.g create a figure, create a plotting area in a figure, plots some lines in a plotting area, decorates the plot with lables, etc.

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,2])
plt.ylabel("Number")                   
plt.xlabel("Indices")
plt.title("Basic plot")
plt.show()
```

If we provide a single list or array to the plot command, matplot assume that it is a sequence of y values and automatically generates the x values for us.Since the python range start with 0 the default x vector has the same length as y but start with 0. Hence the x data are [0,1,2,3].

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("Number")
plt.ylabel("square of Number")
plt.grid()                          # set grid
plt.show
```

for every x,y pair of argument there is an optional third argument which is the format string that indicates the color and line type of the plot.

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.xlabel("Number")
plt.ylabel("square of Number")
plt.grid()                          # set grid
plt.show
```

If matplotlib were limited to working with lists, it would be fairly useless for numeric processing. Generally we will use numpy array. In fact all sequences are converted to numpy arrays internally.

```python
import numpy as np
```

```python
t = np.arange(0.,5.,0.2)


plt.plot(t,t**2,"r--",label='^2')
plt.plot(t,t**2.2,"gs",label='^2.2')
plt.plot(t,t**2.5,"y*",label='^2.5')
plt.grid()
plt.legend()                 # add legend based on line labels
plt.show()
```

# Controlling  line properties

using keyward argument

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5)        # to change line width
plt.grid()
plt.show()
```

use the step()

```python
x1=[1,2,3,4]
y1=[1,4,9,16]
x2=[1,2,3,4]
y2=[2,4,6,8]
lines=plt.plot(x1,y1,x2,y2)

# use keyword args
plt.setp(lines[0],color='r',linewidth=2.0)
plt.grid()
```

#  Working with multiple figure and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)


plt.figure(1)

# The subplot() command specifies numrows, numcols
# fignum where fignum ranges from 1 to numrows*numcols.

plt.subplot(211)        # split the figure 1 into 2 row and 1 column and in 1st plot
plt.grid()
plt.plot(t1,f(t1),"b-")

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),"r--")
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1,2,3])
plt.subplot(212)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(211)
plt.title("Simple as 1,2,3")
plt.show()
```
